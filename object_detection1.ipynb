```
detector = ObjectDetection()
detector.setModelTypeAsTinyYOLOv3()
detector.setModelPath('/content/drive/My Drive/yolo_mammo/yolo-tiny.h5')
detector.loadModel()
detections = detector.detectObjectsFromImage(input_image="/content/download.jpg", output_image_path="/content/drive/My Drive/m.jpg", minimum_percentage_probability=30)

for eachObject in detections:
    print(eachObject["name"] , " : ", eachObject["percentage_probability"], " : ", eachObject["box_points"] )
    print("--------------------------------")
```
